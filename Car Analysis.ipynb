```python
def analyze_dataset(data):
    print('\nThe first rows are : ')
    print(data.head())
    print('\nThe data shape is : ')
    print(data.shape)
    print('\nThe data columns are : ')
    print(data.columns)
    
    print('\n\nThe data columns are : ')
    print(data.dtypes)
    columns = data.columns
    for col in columns:
        print('\n\nAll the unique values for {} are : '.format(col))
        print(data[col].unique())
        
    print('\n\nThe number of unique data columns are : ')
    print(data.nunique())
    print('\n\nThe number of none Null datas are : ')
    print(data.count())
    
    
    print('\n\nThe number of Null datas are : ')

    print(data.isnull().sum())
    print('\n\nThe Info : ')
    print(data.info())
    
    
```

## We have to do some preprocessing task on this dataset.

## Fill Null data

```python
least_occuring_value = Data['Make'].value_counts().idxmin()
Data['Make'].fillna(least_occuring_value, inplace=True)

least_occuring_value = Data['Model'].value_counts().idxmin()
Data['Model'].fillna(least_occuring_value, inplace=True)

least_occuring_value = Data['Type'].value_counts().idxmin()
Data['Type'].fillna(least_occuring_value, inplace=True)

least_occuring_value = Data['Origin'].value_counts().idxmin()
Data['Origin'].fillna(least_occuring_value, inplace=True)

least_occuring_value = Data['DriveTrain'].value_counts().idxmin()
Data['DriveTrain'].fillna(least_occuring_value, inplace=True)

Data['EngineSize'].fillna(Data['EngineSize'].mean(), inplace = True)

Data['Cylinders'].fillna(Data['Cylinders'].mean(), inplace = True)

Data['Horsepower'].fillna(Data['Horsepower'].mean(), inplace = True)

Data['MPG_City'].fillna(Data['MPG_City'].mean(), inplace = True)

Data['MPG_Highway'].fillna(Data['MPG_Highway'].mean(), inplace = True)

Data['Weight'].fillna(Data['Weight'].mean(), inplace = True)

Data['Wheelbase'].fillna(Data['Wheelbase'].mean(), inplace = True)

Data['Length'].fillna(Data['Length'].mean(), inplace = True)
```

```python
Data['MSRP'].fillna('-11', inplace = True)
Data['Invoice'].fillna('-11', inplace = True)
Data[Data['MSRP']=='-11']
```

```python
Data['MSRP'] = Data['MSRP'].str.replace('$', '').str.replace(',', '').astype(int)
Data['Invoice'] = Data['Invoice'].str.replace('$', '').str.replace(',', '').astype(int)
Data[Data['MSRP']==-11]
```

```python
Data['MSRP']=Data['MSRP'].replace(-11,Data['MSRP'].mean())
Data['Invoice']=Data['Invoice'].replace(-11,Data['Invoice'].mean())
```
